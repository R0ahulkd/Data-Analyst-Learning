```python
x = [1,2,3,4,5]
y = [45,34,56,23,45]
plt.subplot(2,2,1) #rows,columns,chart number
plt.plot(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]
plt.subplot(2,2,2) #rows,columns,chart number
plt.bar(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]
plt.subplot(2,2,3)
plt.stackplot(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]
plt.subplot(2,2,4) #rows,columns,chart number
plt.plot(x,y)
plt.suptitle("Employee Data")
plt.show()
```
