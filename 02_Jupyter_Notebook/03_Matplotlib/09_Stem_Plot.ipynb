```python
import matplotlib.pyplot as plt
import pandas as pd

x = [10,40,20,40,20,40,20,40,60,70,50,40]

plt.stem(x)
plt.show()
```

```python
x = [10,40,20,40,20,40,20,40,60,70,50,40]

plt.stem(x, linefmt = "--")
plt.show()
```

```python
x = [10,40,20,40,20,40,20,40,60,70,50,40]

plt.stem(x, linefmt = "--", markerfmt = "D")
plt.show()
```

```python
x = [10,40,20,40,20,40,20,40,60,70,50,40]

plt.stem(x, linefmt = "--", markerfmt = "D", bottom = 5)
plt.show()
```

```python
x = [10,40,20,40,20,40,20,40,60,70,50,40]

plt.stem(x, linefmt = "--", markerfmt = "D", bottom = 5, orientation = "horizontal")
plt.show()
```
