```python
import numpy as np

a = np.array([20,40,60,80])
print(np.append(a,90))
```

```python
a = np.array([[20,40],[60,80]])
print(np.append(a,[90,100])) # It gets converted in 1 dimension then append new element
```

```python
a = np.array([20,40,60,80])
print(np.insert(a,1,50)) # Insert element in required index (array,index,value)
```

```python
a = np.array([[20,40],[60,80]])
print(np.insert(a,1,50))
```

```python
a = np.array([[20,40],[60,80]])
print(np.insert(a,1,[50,60], axis = 0))
```

```python
a = np.array([[20,40],[60,80]])
print(np.insert(a,1,[50,60], axis = 1))
```

```python
a = np.array([[20,40],[60,80]])
print(np.insert(a,1,[50], axis = 0)) # print 50 two times because it is in matrix form
```

```python
a = np.array([[20,40],[60,80]])
print(np.insert(a,[1,2],[50], axis = 0))
```

```python
a = np.array([[20,40],[60,80]])
print(np.delete(a,1)) # First it converts in 1D then index 1 element deleted
```

```python
a = np.array([[20,40],[60,80]])
print(a)
print(np.delete(a,1, axis = 0)) # Delete row in index 1
```

```python
a = np.array([[20,40],[60,80]])
print(a)
print(np.delete(a,1, axis = 1)) # Delete column in index 1
```
