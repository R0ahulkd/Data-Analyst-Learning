```python
a = [30,40,50]
b = [5,5,3]
print(a+b) # add 2 list
```

```python
import numpy as np

# Concatenate
arr1 = np.array([30,40,50])
arr2 = np.array([5,5,3])
print(np.concatenate([arr1,arr2]))
```

```python
print(np.concatenate([arr1,arr2],axis=0)) # same output as becasue 1 dimension
```

```python
arr1 = np.array([[30,40],[50,10]])
arr2 = np.array([[5,5],[3,3]])
print(np.concatenate([arr1,arr2]))
```

```python
print(np.concatenate([arr1,arr2],axis=0))
print(np.concatenate([arr1,arr2],axis=1)) # Now there is 2 dimension arrays so output will change
```

```python
print(np.hstack([arr1,arr2])) # Horizontal Concatenation
```

```python
print(np.vstack([arr1,arr2])) # Vertical Concatenation
```

```python
#Split

a = np.array([20,40,30,40,10,20])
print(np.array_split(a,3)) # It splits the array in 3 arrays
```

```python
a = np.array([20,40,30,40,10,20])
b = (np.array_split(a,3))
print(b[1]) # It prints array 1 as index is given
```
