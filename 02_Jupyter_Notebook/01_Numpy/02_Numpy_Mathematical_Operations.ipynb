```python
import numpy as np
```

```python
arr1 = np.array([30,40,60,20,10])
arr2 = np.array([30,20,10,20,20])
print(arr1 + arr2) # It adds value of arr1 in arr2 like 30+40 and 40+20
```

```python
print(np.add(arr1,arr2)) # Same as print(arr1 + arr2)
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1 + arr2)
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1 - arr2)
```

```python
print(np.subtract(arr1,arr2)) # Same as print(arr1 - arr2)
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1 * arr2)
```

```python
print(np.multiply(arr1,arr2)) # Same as print(arr1 * arr2)
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1 / arr2)
```

```python
print(np.divide(arr1,arr2)) # Same as print(arr1 / arr2)
```

```python
arr1 = np.array([3,4,2,10])
arr2 = np.array([2])
print(np.power(arr1,arr2)) # It prints power in 3 of all elements of arr1
```

```python
arr1 = np.array([9,16,4,100])
print(np.sqrt(arr1)) # Find square root
print(arr1.dtype)
```
