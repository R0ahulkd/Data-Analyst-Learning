```python
import numpy as np
baked_food = [200,300,150,130,200,280,170,188]

a = np.array(baked_food)
print(np.mean(baked_food)) # Sum of all values divided by number of values
```

```python
print(np.median(baked_food)) # Gives central value (After Sorting)
```

```python
import statistics as stats # Mode is available in statisctic library
print(stats.mode(baked_food)) #  Most occuring value
```

```python
print(np.std(baked_food)) # Gives standard deviation
```

```python
print(np.var(baked_food)) # Gives Variance (Square of Standard deviation)
```

```python
# -1 represent inversely proportional relationship
# 1 represents proportional relationship
# 0 mean no relationship
tobacco_consumption = [30,50,10,30,50,40]
deaths = [100,120,70,100,120,112]

print(np.corrcoef([tobacco_consumption,deaths])) # Coefficient of correlation
```
