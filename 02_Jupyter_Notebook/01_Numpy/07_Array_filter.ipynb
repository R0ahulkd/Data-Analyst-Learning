```python
import numpy as np

ar= np.array([20,30,40,50])
fa = [True, False, True, False]
new = ar[fa]
print(new)
# It passes the value of ar in new where fa is True
```

```python
ar = np.array([20,30,40,50])
fa = ar>35
new = ar[fa]
print(new)
# new stores value greater than 35 from ar
```
