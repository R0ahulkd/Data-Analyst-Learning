```python
import pandas as pd

data = pd.read_csv("company1.csv")
print(data)
```

```python
print(data.isnull())
```

```python
print(data.isnull().sum())
```

```python
print(data.dropna()) # It drops all columns with null values
```

```python
print(data)
```

```python
import numpy as np
print(data.replace(np.nan, "hi")) # Replace all nan values with hi
```

```python
import numpy as np
print(data.replace(np.nan, 30000))
```

```python
# Replace nan values in salary as 30000 only in salary column
data["salary"] = data["salary"].replace(np.nan, 30000)
print(data)
```

```python
# Replace salary with mean salary of company for nan values
print(data)
average = data["salary"].mean()
print(average)
data["salary"] = data["salary"].replace(np.nan, average)
print(data)
```

```python
print(data["salary"].mean())
```

```python
print(data)
```

```python
print(data.fillna(method = "bfill")) # b means backfill use to fill with back value like Nan is filled by F which is fill in next column
```

```python
data.bfill() # Same as print (data.fillna(method = "bfill"))
```

```python
data.ffill() # f means forwardfill use to fill with back value like Nan is filled by M which is fill in next column
```
