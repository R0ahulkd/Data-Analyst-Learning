```python
import pandas as pd

dict = {"Fruits":["Mango","Apples","Banana","Papaya"],
       "Price":[100,150,50,35],
       "Quantity":[15,10,10,3]}

df1 = pd.DataFrame(dict)
print(df1)
```

```python
df2 = df1.copy()
print(df2)
df2.loc[0,"Price"] = 120
df2.loc[1,"Price"] = 175
df2.loc[2,"Price"] = 30
df2.loc[0,"Quantity"] = 12
df2.loc[1,"Quantity"] = 15
df2.loc[2,"Quantity"] = 5
print(df2)
```

```python
print(df1.compare(df2)) # Show differences between df1 and df2
```

```python
print(df1.compare(df2,align_axis = 0))
```

```python
print(df1.compare(df2,keep_equal = True)) # Compare df1 and df2 and show both equal and different values
```

```python
# Compare df1 and df2 while keeping the original shape (rows/columns) using NaN for equal values
print(df1.compare(df2,keep_shape = True))
```

```python
print(df1.compare(df2,keep_shape = False))
```
