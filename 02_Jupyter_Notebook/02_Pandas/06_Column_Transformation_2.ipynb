```python
# Create a new column by taking first 3 letters of each month
# Used .map() to apply a function on each cell of the Months column
# Result: short month names like Jan, Feb, Mar


data = {"Months":["January","February","March","April"]}

a = pd.DataFrame(data)
print(a)

def extract(value):
    return value[0:3]

a["Short_Months"] = a["Months"].map(extract)
print(a)
```
