```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
gp = data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```
