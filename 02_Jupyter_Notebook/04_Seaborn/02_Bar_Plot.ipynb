```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("titanic")
print(data)
```

```python
data = sns.load_dataset("tips")
print(data)
```

```python
sns.barplot(data = data, x = "day", y = "tip")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", estimator = "median")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", estimator = "sum")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "Blues")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "viridis")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "GnBu")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "spring")
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "spring", order = ["Sun","Sat","Fri","Thur"])
plt.plot()
```

```python
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "spring", order = ["Sun","Sat","Fri","Thur"], errorbar = ("ci",0))
plt.plot()
```
