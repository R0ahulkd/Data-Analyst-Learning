```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#KDE (Kernel Density Estimation)

data = sns.load_dataset("tips")
print(data)
```

```python
sns.kdeplot(data = data, x = "total_bill")
plt.show()
```

```python
sns.histplot(data = data, x = "total_bill")
plt.show()
```

```python
sns.kdeplot(data = data, x = "total_bill", hue = "day")
plt.show()
```

```python
sns.kdeplot(data = data, x = "total_bill", hue = "day", multiple = "stack")
plt.show()
```

```python
sns.kdeplot(data = data, x = "total_bill", hue = "day", multiple = "fill")
plt.show()
```

```python
sns.kdeplot(data = data, x = "total_bill", hue = "day", multiple = "fill", palette = "viridis")
plt.show()
```
