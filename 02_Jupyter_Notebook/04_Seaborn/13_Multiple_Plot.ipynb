```python
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset("tips")
print(data)
```

```python
a = sns.FacetGrid(data, col = "day")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "smoker")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "smoker", height = 2)
plt.show()
```

```python
a = sns.FacetGrid(data, col = "smoker", height = 2)
a.map(sns.barplot,data = data, x = "day", y = "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "smoker", height = 2)
a.map(sns.scatterplot,data = data, x = "day", y = "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "time", height = 2)
a.map(sns.scatterplot,data = data, x = "day", y = "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "time", height = 2, hue = "sex")
a.map(sns.barplot,data = data, x = "day", y = "tip")
plt.show()
```

```python
a = sns.FacetGrid(data, col = "time", height = 5, hue = "sex")
a.map(sns.barplot,data = data, x = "day", y = "tip")
plt.show()
```
