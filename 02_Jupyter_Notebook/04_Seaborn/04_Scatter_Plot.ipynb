```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip")
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "smoker")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "time")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size = "size")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size = "size", marker = "s")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size = "size", marker = "d")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size = "size", palette = "spring")
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "day", size = "size", palette = "viridis")
plt.show()
```

```python
data = pd.read_excel("C:/Users/Rahul Kumr/Documents/Github/Data-Science-Learning/02_Jupyter_Notebook/02_Pandas/Employee Sample Data.xlsx")
df = pd.DataFrame(data)
print(df)
```

```python
sns.scatterplot(data = df, x = df["Age"], y = df["Annual Salary"], size = df["Bonus %"])
plt.show()
```

```python
sns.scatterplot(data = df, x = df["Age"], y = df["Annual Salary"], size = df["Bonus %"], hue = "Gender")
plt.show()
```

```python
sns.scatterplot(data = df, x = df["Age"], y = df["Annual Salary"], size = df["Bonus %"], hue = "Department")
plt.show()
```

```python
sns.scatterplot(data = df, x = df["Age"], y = df["Annual Salary"], size = df["Bonus %"], hue = "Department")
plt.legend(bbox_to_anchor = (0.8,1.2,0.8,1.2)) # (x-axis, y-axis, width, Height)
plt.show()
```

```python
sns.scatterplot(data = df, x = df["Age"], y = df["Annual Salary"], size = df["Bonus %"], hue = "Department")
plt.legend(bbox_to_anchor = (0.2,0,1.2,1.2))
plt.show()
```
